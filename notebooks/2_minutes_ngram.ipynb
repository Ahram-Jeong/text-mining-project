```python
from ekonlpy.sentiment import MPCK
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
# 크롤링 결과물 불러오기
df = pd.read_csv("../data/minutes_df.csv")
call_df = pd.read_csv("../data/call_rate_diff.csv")
df
```

```python
# contents가 없는 row 삭제 (결측치 제거)
df = df.dropna()
df
```

```python
# 의사록 어조 분석용
df.to_csv("../data/minutes_contents.csv", index = False, encoding = "utf-8")
```

## 데이터 전처리
1. 토큰화
2. n-gram화

```python
# 텍스트 전처리
mpck = MPCK()

# 1. 토큰화
df["tokens"] = df["contents"].astype(str).apply(mpck.tokenize)
# 2. ngramize
df["5grams"] = df["contents"].astype(str).apply(mpck.tokenize).apply(mpck.ngramize)
df
```

### 토큰화 결과에 대한 필터링

```python
# 품사 필터링 : 명사(NNG), 형용사(VA, VAX), 부사(MAG), 동사(VA), 부정(MAG)
def filter_pos(tokens) :
    pos = ["NNG", "VA", "VAX", "MAG", "VV"]
    filtered_tokens = []
    for token in tokens :
        word, tag = token.split("/")
        if tag in pos :
            filtered_tokens.append(token)
    return filtered_tokens
    
# str 길이 2 이상만 필터링
def filter_len(tokens) :
    return [token for token in tokens if len(token.split("/")[0]) > 1]
```

#### 토큰화와 n-gram화 컬럼 병합

```python
# 필터링 함수 호출 
# tokens 컬럼 필터링
df["tokens"] = df["tokens"].apply(filter_pos).apply(filter_len)
# tokens + 5grams -> ngrams 컬럼 생성
df["ngrams"] = df["tokens"] + df["5grams"]
df
```

```python
# 콜 금리 DataFrame과 병합
call_m_df = pd.merge(df, call_df, left_on = "date", right_on = "date", how = "inner")
call_m_df
```

```python
call_m_df.info()
```

```python
call_m_df = call_m_df.reindex(columns = ["date", "title", "contents", "ngrams", "Labeling"])
call_m_df
```

## ngrams의 unique value 추출
- ngrams 컬럼의 모든 요소들을 list에 추가
- list를 set으로 변환하여 중복 값 제거

```python
# ngrams 컬럼의 요소들을 list로 추출
n_list = call_m_df["ngrams"].values.tolist()
n_list
```

```python
# 2차원 list를 1차원으로 flatten 하는 함수
def flatten_list(ng_list) :
    flat_list = []
    for sublist in ng_list :
        for i in sublist :
            flat_list.append(i)
    return flat_list

# 함수 호출
result = flatten_list(n_list)
len(result), result
```

```python
# list -> set으로 변환하여 중복 값 제거
set_lst = set(result)
len(set_lst), set_lst
```

```python
# ngrams 컬럼의 각 요소들을 list에서 str로 변환하여 str_ngrams 컬럼 생성
call_m_df["str_ngrams"] = None
for i in range(len(call_m_df["ngrams"])) :
    call_m_df["str_ngrams"][i] = ", ".join(call_m_df["ngrams"][i])
call_m_df
```

```python
# 필요한 컬럼들로 reindex
result_df = call_m_df.reindex(columns = ["date", "title", "str_ngrams", "Labeling"])
result_df = result_df.rename(columns = {"str_ngrams" : "ngrams"}) # str_ngrams 리네임
result_df
```

```python
# 의사록 전처리 결과물 (ngrams) 다운로드
result_df.to_csv("../data/minutes_ngrams.csv", encoding = "utf-8")
```

## 금리 사전 구축

```python
# 의사록 전처리 결과물에서 Labeling 결과 값에 따라 Up / Down 분류한 데이터프레임 생성
label_up = result_df[result_df.Labeling == "Up"]
label_down = result_df[result_df.Labeling == "Down"]
label_up.reset_index(drop = True, inplace = True)
label_down.reset_index(drop = True, inplace = True)
print(len(label_up))
print(len(label_down))
```

#### 사전 구축을 위한 ngram 빈도 수 측정

```python
word_cnt_df = {}
for i in set_lst :
    word_cnt_df[i] = {"up" : 0, "down" : 0}
```

```python
for ngram_list in tqdm(label_up.ngrams):
    ngram = ngram_list.split(", ")
    for i in ngram :
        word_cnt_df[i]["up"] += 1
        
for ngram_list in tqdm(label_down.ngrams):
    ngram = ngram_list.split(", ")
    for i in ngram :
        word_cnt_df[i]["down"] += 1
```

```python
result_cnt_df = pd.DataFrame.from_dict(word_cnt_df, orient ="index", columns = ["up", "down"])
result_cnt_df["total"] = result_cnt_df.up + result_cnt_df.down
result_cnt_df
```

```python
# total >= 15으로 cnt_df 데이터프레임 생성
cnt_df = result_cnt_df[result_cnt_df["total"] >= 15]

# Laplace Smoothing
cnt_df["up"] += 1
cnt_df["down"] += 1
cnt_df["total"] = cnt_df["up"] + cnt_df["down"]
cnt_df
```

## Polarity Score (극성 점수) 구하기

```python
cnt_df["polar_score"] = (cnt_df["up"]/cnt_df["up"].sum()) / (cnt_df["down"]/cnt_df["down"].sum())
cnt_df["polarity"] = cnt_df["polar_score"].apply(lambda x: "Hawkish" if x > 1.3 else ("Dovish" if x < 0.76 else ""))
cnt_df
```

```python
# 결측치 제거
cnt_df = cnt_df.replace("", np.nan)
cnt_df = cnt_df.dropna()
cnt_df
```

### 금리 상승 사전 생성

```python
pos_dict = cnt_df[cnt_df["polarity"] == "Hawkish"]
pos_dict.index.name = "word"
pos_dict.to_csv("../data/pos_dict.csv", encoding = "utf-8")
pos_dict
```

### 금리 하락 사전 생성

```python
neg_dict = cnt_df[cnt_df["polarity"] == "Dovish"]
neg_dict.index.name = "word"
neg_dict.to_csv("../data/neg_dict.csv", encoding = "utf-8")
neg_dict
```
